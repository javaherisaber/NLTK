## 14 - Define a function supergloss(s) that takes a synset s as its argument and returns a string consisting of the concatenation of the definition of s, and the definitions of all the hypernyms and hyponyms of s.

```python
from nltk.corpus import wordnet as wn
```

```python
def supergloss(s):
    df0 = wn.synset(s).definition()
    output = s + ' : ' + df0 + '\n'
    output += 'hypernyms of ' + s + ' :: ' + '\n'
    for hyper in wn.synset(s).hypernyms():
        df1 = wn.synset(hyper.name()).definition()
        output += hyper.name() + ' : ' + df1 + '\n'
    output += 'hyponyms of ' + s + ' ::: ' + '\n'
    for hypo in wn.synset(s).hyponyms():
        df2 = wn.synset(hypo.name()).definition()
        output += hypo.name() + ' : ' + df2 + '\n'
    return output
```

```python
print(supergloss('car.n.01'))
```

## 18 - Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords.

```python
from nltk.corpus import stopwords
from nltk import bigrams
from nltk import FreqDist
```

```python
def most_meaningful_bigrams(text):
    en_stop_words = stopwords.words('english')
    without_stop_words = [w for w in text if w.lower() not in en_stop_words]
    bigrams_freq = FreqDist(bigrams(without_stop_words))
    return bigrams_freq.most_common(50)
```

```python
from nltk.corpus import genesis
```

```python
text = genesis.words('english-kjv.txt')
```

```python
most_freq = most_meaningful_bigrams(text)
```

## 25 - Define a function find_language() that takes a string as its argument, and returns a list of languages that have that string as a word. Use the udhr corpus and limit your searches to files in the Latin-1 encoding.

```python
from nltk.corpus import udhr
```

```python
def find_language(string):
    latin1_fileids = [fileid for fileid in udhr.fileids() if fileid.endswith('-Latin1')]
    language_used = [fileid for fileid in latin1_fileids if string.lower() in [w for w in udhr.words(fileid)]]
    return language_used
```

```python
find_language('cura')
```

## 26 - What is the branching factor of the noun hypernym hierarchy? I.e. for every noun synset that has hyponyms  or children in the hypernym hierarchy  how many do they have on average? You can get all noun synsets using wn.all_synsets('n').

```python
from nltk.corpus import wordnet as wn
```

```python
noun_synsets = wn.all_synsets('n')
children_count = 0
item_count = 0
for item in noun_synsets:
    hypo_len = len(item.hyponyms())
    if(hypo_len):
        children_count += hypo_len
        item_count += 1
factor = children_count / item_count
print('Branching Factor for nouns is : ', factor)
```

## 27 - The polysemy of a word is the number of senses it has. Using WordNet, we can determine that the noun dog has 7 senses with:  len(wn.synsets('dog', 'n')). Compute the average polysemy of nouns, verbs, adjectives and adverbs according to WordNet.

```python
from nltk.corpus import wordnet as wn
```

```python
pos = ['n','v','a','r']
for item in pos:
    word_count=0
    word_sum=0
    for word in wn.words():
        word_sum += len(wn.synsets(word,item))
        word_count += 1
    print('POS (',item,') is :',word_sum/word_count)
```

```python
def similarity(word1, word2):
    synset1 = wn.synset(word1 + '.n.01')
    synset2 = wn.synset(word2 + '.n.01')

    return synset1.path_similarity(synset2)
```
