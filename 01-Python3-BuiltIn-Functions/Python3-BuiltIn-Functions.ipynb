## 51- class property(fget=None, fset=None, fdel=None, doc=None)

```python
class Test:
    def __init__(self):
        self._t = 1

    @property
    def t(self):
        """I'm the 't' property."""
        return self._t

    @t.setter
    def t(self, value):
        self._t = value

    @t.deleter
    def t(self):
        del self._t
t1 = Test()
t1.t = 2
t1.t
```

## 63- super([type[, object-or-type]])

```python
class A:
    def met(self,arg):
        self.arg = arg +1

class B(A):
    def met(self,arg):
        super(B, self).met(arg)
        self.arg = arg +2
b = B()
b.met(2)
b.arg
```
