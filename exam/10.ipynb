```python
from nltk.corpus import reuters
from nltk import FreqDist
fd = FreqDist(reuters.words())
lst = [s for s in fd.most_common()]
print(lst)
```

```python
from nltk import ConditionalFreqDist

cond = ConditionalFreqDist(
  (categ, word)
  for categ in reuters.categories()
  for word in reuters.words(categories=[categ]))
```
